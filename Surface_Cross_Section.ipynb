```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import ipywidgets as widgets

@widgets.interact(y_pointer=(0,28,1))
def slice_demo(y_pointer = 0.0):

    N = 29;
    x,y = np.linspace(-1,1, N*2), np.linspace(-1,1, N)
    X,Y = np.meshgrid(x,y)
    F = lambda X,Y : np.sin(8*X)/(1+5*(X**2+Y**2))
    Z = F(X,Y)

    Ys = Y*0
    Ys[:,:] = .3
    Ys[y_pointer] = 1
    #Ys[20] = .98
    cmap = plt.cm.viridis
    xx,zz = np.meshgrid(x,y)
    y_plane = np.linspace(-1,1, N)
    yy = np.full_like(Y, y_plane[y_pointer])
    
    plt.figure(1,figsize = (12,6))
    ax = plt.subplot(121, projection='3d')
    ax.plot_surface(X, Y, Z, facecolors=cmap(Ys),alpha=0.5)
    ax.plot_surface(xx, yy, zz, alpha=0.5, color='r')
    plt.subplot(122)
    plt.plot(x,Z[y_pointer,:], color='r')

    plt.show()
```
